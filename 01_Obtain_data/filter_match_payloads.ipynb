```python
import requests
import json
import pandas as pd
import time 
import random
import os
from IPython.display import display, clear_output
```

```python
payload_directory = "Data/Raw/Match_payloads"
files =  [file for file in os.listdir(payload_directory) if file.endswith(".csv")]
output_dir = "Data/Processed/Match_payloads"
os.makedirs (output_dir,exist_ok=True)
print(files)
```

```python
def filter_singles(df: pd.DataFrame) -> pd.DataFrame:
    condition = df['subEventType'].str.contains('Singles', na=False)
    singles_df = df[condition].copy()
    singles_df.reset_index(drop=True, inplace=True)
    event_id = df["eventId"][1]
    print(f"{len(df)} matches found for event ID:{event_id} keeping {len(singles_df)} singles matches ")
    return singles_df
```

```python
for file in files: 
    
    file_path = f"{payload_directory}/{file}"
    
    payloads_df = pd.read_csv(file_path) 
    
    singles_df = filter_singles(payloads_df)
   
    event_id = singles_df["eventId"].iloc[1]
    
    output_file = f"{output_dir}/{event_id}_singles_payloads.csv"
    print(f"Saving to {output_file}")
    singles_df.to_csv(output_file,index=False)
    
    


    
   
```
