```python
import pandas as pd
import os
from typing import List, Union
```

```python
"""
This script reads and filters events data that have been scraped by year. 
All desired events are saved into a master file.
"""

# Configuration

# Define directory where events by year files are stored as csv.
events_directory = "./Data/Raw/Events"

# Output directory where the master file will be saved
output_directory="./Data/Processed/Events/Test"

"""
Manually Filtering Events by Type:

Only Keeping WTT Series Mainline Events (Feeder, Contender, Star Contender, Champions, Smash, Tour Finals),
As well as singles World Cup and World Championship.
"""

# EventTypeIds to keep as explained above: 
# Determined by manually parsing the events in the events_files.
keep_ids = [29,34,35,37,38,64,65,75,81,85,91,95]

# Strings used to filter out team and qualifying event.
# "qualif" used to encapsulate "qualifying", "qualifiers", "qualification" etc
remove_strings = ["qualif", "team"]


# This can be used to rename events after filtering for increased clarity
# Passed into the rename_events function.
name_map = {
    "Singles World Cup": "World Cup",
    "WTT Cup Finals": "WTT Finals",
    "WTTC": "World Championship"
}
```

```python
def collate_raw_events (directory: str) -> pd.DataFrame:
    """
    Loads all individual event CSV files from the specified directory and compiles them 
    into a single DataFrame.
    """
    all_events_list = [] 
    print("--- 🟢 Combining Raw Event Files 🟢 ---")

    # Iterate through csv files in search directory
    for file in os.listdir(directory):
        if file.endswith(".csv"):
            
            print(f"Reading file: {file}")
    
            # using os.path.join to create the full file path from the directory and filename
            # read the file, convert to DF, ana store in all_events_list container.
            
            full_path = os.path.join(directory, file)
            df = pd.read_csv(full_path)
            all_events_list.append(df)

    if not all_events_list:
        print(f"❌ Error: No CSV files found in {directory}.")
        # Return blank DataFrame if no data found
        return pd.DataFrame() 

    all_events_df = pd.concat(all_events_list, ignore_index=True)

    return all_events_df

    
def filter_selected_events(df: pd.DataFrame,
                           keep_ids: List[int],
                           remove_strings: List[str]
                           ) -> pd.DataFrame:
    """
    Filters a dataframe of events data in order to keep only the desired events as specified by the inputs.
    Here: WTT mainline events, Olympics, World Cup, World Championships - singles only.
    Also renames event types for clarity based on name_map
    """
    print(f"\n--- 🟢 Filtering from {len(df)} Events 🟢 ---")

    # Firstly filter events with the desired EventTypeIds (i.e of the desired event types)
    filtered_df = df[df["EventTypeId"].isin(keep_ids)].copy()

    # here "|" denotes "OR" for regex pattern. If the event name has any of these terms it will be removed
    pattern = "|".join(remove_strings)
    
    # create a mask for conditions to filter the DF - here select all event names with those strings.
    mask = filtered_df["EventName"].str.contains(pattern,case=False)

    # use filter condition ~mask to select entries that DO NOT contain the filtered strins
    filtered_df = filtered_df[~mask] 

    print(f"Filtered events: {len(df)} -> {len(filtered_df)} ")

    
    # Remove any duplicate listings.
    filtered_df.drop_duplicates(subset=["EventId"], keep = "first", inplace = True)
    
    print(f"Removed duplicates, final event count = {len(filtered_df)}")
    

   

def standardize_event_names(df: pd.DataFrame, name_map: dict) -> pd.DataFrame:
    """
    Changes event names from the original event list to simpler names for Clarity.
    e.g., ['Singles World Cup', "WTTC"] to ["World Cup", "World Championships"]
    
    World Cup (newer, annual) and World Championship (older, biannnual) are often confused.
    Name changes can be specified in name_map variable.
    
    """
    # Use .loc to explicitly change names of input df.
    df["EventType"] = df["EventType"].replace(name_map)
    return df
```

```python
if __name__ == "__main__":
    
    # Create the output directory if it does not exist.
    # Note: Using exist_ok=True prevents an error if the directory already exists.
    os.makedirs(output_directory, exist_ok=True) 

    print("---🚀 Starting WTT Event Processing 🚀---")
    
    # 1. Load and combine all raw data
    raw_events_df = collate_raw_events(directory=events_directory)
    
    if raw_events_df.empty:
        print("--- ❌ Processing failed: No raw data loaded. Check the input directory. ---")
    else:
        # 2. Filter the events to keep only the mainline singles tournaments
        filtered_df = filter_selected_events(
            df=raw_events_df,
            keep_ids=keep_ids,
            remove_strings=remove_strings
        )
        
        # 3. Standardize the event names for consistency
        curated_df = standardize_event_names(
            df=filtered_df,
            name_map=name_map
        )
        
        # 4. Final Sort (Optional: ensures a consistent output order)
        curated_df = curated_df.sort_values(["EventType", "StartDateTime"])
        
        # 5. Save the final curated event list
        master_path = os.path.join(output_directory, "curated_events_master.csv")
        curated_df.to_csv(master_path, index=False)
        
        print(f"\nTotal events saved to master file: {len(curated_df)}")
        print(f"✅ Master event list successfully saved to '{master_path}'")
        

    print("\n---🟢 Processing finished. 🟢---")
```
