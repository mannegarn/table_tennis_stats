```python
import requests
import json
import pandas as pd
import os

def get_events_by_year(start_year,end_year=2020, end_year=2025 output_dir="./"):
    """
    Fetches all the WTT events and event info for specified year_rangesand saves as csv.
    Args:
        year (int or str): The year to fetch events for.
        output_dir (str): The directory to save the raw data in.
        
    Returns:
        bool: True if successful, False otherwise.
    """
    url = 'https://wtt-website-api-prod-3-frontdoor-bddnb2haduafdze9.a01.azurefd.net/api/eventcalendar'

    headers = {
        'accept': 'application/json, text/plain, */*',
        'origin': 'https://www.worldtabletennis.com',
        'referer': 'https://www.worldtabletennis.com/',
        'user-agent': 'Mozilla/5.0 (Linux; Android 11.0; Surface Duo) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/140.0.0.0 Mobile Safari/537.36'
    }

    data = {
        "custom_filter": f"[{{\"name\":\"StartDateTime\",\"value\":{start_year},\"custom_handling\":\"multimatch_year_or_filter\",\"condition\":\"or_start\"}},{{\"name\":\"FromStartDate\",\"value\":{end_year},\"custom_handling\":\"multimatch_year_or_filter\",\"condition\":\"or_end\"}}]"
    }

    try:
        print(f"Fetching events for the years {start_year}-{end_year}")
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()

        response_data = response.json()
        events_list = response_data[0].get('rows')

        if not (events_list and isinstance(events_list, list)):
            print(f"No event data found for range {start_year} - {end_year}.")
            return False

        # --- Create DataFrame with ALL columns from the API response ---
        df = pd.DataFrame(events_list)
        print(f"Found {len(df)} events. Saving to raw CSV...")

        # Create the output directory if it doesn't exist
        os.makedirs(output_dir, exist_ok=True)
        filename = f'{output_dir}/raw_events_{start_year}{end_year}.csv'
        
        df.to_csv(filename, index=False)
        print(f"✅ Raw data successfully saved to {filename}")
        return True

    except requests.exceptions.HTTPError as err:
        print(f"❌ HTTP Error: {err}")
    except json.JSONDecodeError:
        print(f"❌ Error: The response was not valid JSON.")
    except Exception as err:
        print(f"❌ An unexpected error occurred: {err}")
    
    return False


if __name__ == "__main__":
    # Define the years you want to scrape
   
    
    print("--- Starting WTT Events Scraper ---")
    for year in years_to_scrape:
        save_raw_events_from_year(year)
        print("-" * 30)
    print("--- Scraping finished. ---")
```
