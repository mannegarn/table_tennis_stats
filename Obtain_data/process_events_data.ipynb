```python
import pandas as pd
import os
output_path="./Data/Processed/Events"


directory = "./Data/Raw/Events"
all_events_list = []

print("Reading all CSV files from the directory...")
for file in os.listdir(directory):
    if file.endswith(".csv"):
        print(f" -  Reading -  {file}")
        full_path = os.path.join(directory, file)
        df = pd.read_csv(full_path)
        all_events_list.append(df)

combined_df = pd.concat(all_events_list, ignore_index=True)
combined_df = combined_df.sort_values(["EventTypeId", "StartDateTime"])

combined_df.to_csv(f"{output_path}/all_events.csv", index=False)

print(f"\n✅ All {len(all_events_list)} Events from all years combined and saved to'{output_path}'")
print(f"Total events loaded: {len(combined_df)}")


```

```python
"""
Manually Filtering Events by Type:

Only Keeping WTT Series Mainline Events (Feeder, Contender, Star Contender, Champions, Smash, Tour Finals),
As well as singles World Cup and World Championship.
"""


all_events_df = pd.read_csv("Data/Processed/Events/all_events.csv")

# Filtering by event type
ids_to_keep = [29,34,35,37,38,64,65,75,81,85,91,95]

filtered_df = all_events_df[all_events_df['EventTypeId'].isin(ids_to_keep)]



# Filtering out qualification and team events (to keep things simple)
strings_to_remove = ["qualifi", "team"]
pattern = "|".join(strings_to_remove)

mask_to_remove = filtered_df["EventName"].str.contains(pattern, case=False)
filtered_df = filtered_df[~mask_to_remove].copy()



names_replace= {
    "Singles World Cup": "World Cup",
    "WTT Cup Finals": "WTT Finals",
    "WTTC": "World Championship" }

filtered_df['EventType'] = filtered_df['EventType'].replace(names_replace)



#Now sort by verbose Event Type and save as curated events !
filtered_df = filtered_df.sort_values(["EventType","StartDateTime"])
filtered_df.to_csv(f"{output_path}/curated_events.csv",index=False)
```
