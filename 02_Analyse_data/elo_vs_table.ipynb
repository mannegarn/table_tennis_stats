```python
import sys
import os
import pandas as pd




project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))

if project_root not in sys.path:
    sys.path.append(project_root)
    

from utils.getLatestFiles import get_latest_master_players, get_latest_master_matches, get_latest_master_events
from utils.glickoHelpers import calculate_glicko2_ratings, get_final_player_stats
```

```python
master_players_df = get_latest_master_players()
master_matches_df = get_latest_master_matches()
master_events_df = get_latest_master_events()
master_matches_df = master_matches_df.sort_values(by=["matchDate"])
master_matches_df = master_matches_df
master_matches_df.reset_index(drop=True, inplace=True)
```

```python
ratings_df = calculate_glicko2_ratings(master_players_df, master_matches_df)
```

```python
final_stats_df = get_final_player_stats(ratings_df)
```

```python

women_ids = master_players_df[master_players_df["Gender"] == "F"]
men_ids = master_players_df[master_players_df["Gender"] == "M"]
```

```python
men_final_stats_df = final_stats_df[final_stats_df["playerId"].isin(men_ids["playerId"])]
women_final_stats_df = final_stats_df[final_stats_df["playerId"].isin(women_ids["playerId"])]
```

```python
grippy_tables = ["DHS", "Double Fish", "Yinhe"]
fast_tables = ["Tibhar", "Stiga", "Joola", "Donic"]
grippy_event_ids = master_events_df[master_events_df["TableSponsor"].isin(grippy_tables)]["eventId"].unique()
fast_event_ids = master_events_df[master_events_df["TableSponsor"].isin(fast_tables)]["eventId"].unique()

grippy_matches_df = master_matches_df[master_matches_df["eventId"].isin(grippy_event_ids)]
fast_matches_df = master_matches_df[master_matches_df["eventId"].isin(fast_event_ids)]


fast_ratings_df = calculate_glicko2_ratings(master_players_df, fast_matches_df, info_string="fast tables")
grippy_ratings_df = calculate_glicko2_ratings(master_players_df, grippy_matches_df, info_string="grippy tables")


fast_final_stats_df = get_final_player_stats(fast_ratings_df, info_string="fast tables")
grippy_final_stats_df = get_final_player_stats(grippy_ratings_df, info_string="grippy tables")
```
