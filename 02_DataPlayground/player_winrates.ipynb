```python
def get_country_winrate_diffs(player_stats_df: pd.DataFrame, min_players: int = 1) -> pd.DataFrame:
    """
    Aggregates player-level win rate differences up to the country level.

    Args:
        player_stats_df (pd.DataFrame): The DataFrame from your example, containing
                                        ['CountryName', 'Fast vs Grippy Winrate Diff', 'playerId'].
        min_players (int, optional): The minimum number of players a country must have
                                     to be included in the final report. Defaults to 1.

    Returns:
        pd.DataFrame: A DataFrame indexed by CountryName, showing the average difference
                      and the number of players.
    """
    
    # 1. Group by 'CountryName' and aggregate multiple stats at once
    country_stats_df = player_stats_df.groupby('CountryName').agg(
        # Calculate the mean of the difference column
        Average_WinRate_Diff=('Fast vs Grippy Winrate Diff', 'mean'),
        
        # Count the number of unique players from that country
        PlayerCount=('playerId', 'nunique')
    )
    
    # 2. Apply the minimum player filter
    if min_players > 1:
        country_stats_df = country_stats_df[country_stats_df['PlayerCount'] >= min_players]
        
    # 3. Sort by the average difference (descending)
    country_stats_df = country_stats_df.sort_values(by='Average_WinRate_Diff', ascending=False)
    
    # 4. Round the average for cleaner output
    country_stats_df['Average_WinRate_Diff'] = country_stats_df['Average_WinRate_Diff'].round(2)
    
    return country_stats_df
```

```python
import pandas as pd
import numpy as np

def get_country_winrate_diffs_weighted(player_stats_df: pd.DataFrame, min_players: int = 1) -> pd.DataFrame:
    """
    Aggregates player-level win rate differences up to the country level,
    weighted by the total number of matches each player played.

    Args:
        player_stats_df (pd.DataFrame): The DataFrame from your example, containing
                                        ['CountryName', 'Fast vs Grippy Winrate Diff', 
                                         'TotalMatches (fast tables)', 'TotalMatches (grippy tables)',
                                         'playerId'].
        min_players (int, optional): The minimum number of players a country must have
                                     to be included in the final report. Defaults to 1.

    Returns:
        pd.DataFrame: A DataFrame indexed by CountryName, showing the weighted
                      average difference, total matches, and player count.
    """
    
    # --- Weighted Average Calculation ---
    
    # 1. Create a copy to avoid modifying your original DataFrame
    df_copy = player_stats_df.copy()
    
    # 2. Calculate the "Weight" for each player (their total matches)
    df_copy['PlayerTotalMatches'] = df_copy['TotalMatches (fast tables)'] + df_copy['TotalMatches (grippy tables)']
    
    # 3. Calculate the "Weighted Value" for each player
    # (Player's Diff * Player's Total Matches)
    df_copy['WeightedDiff'] = df_copy['Fast vs Grippy Winrate Diff'] * df_copy['PlayerTotalMatches']
    
    # 4. Group by 'CountryName' and sum up the components
    country_stats_df = df_copy.groupby('CountryName').agg(
        # Get the sum of all weighted diffs
        TotalWeightedDiff=('WeightedDiff', 'sum'),
        
        # Get the sum of all matches (this is our total weight)
        TotalCountryMatches=('PlayerTotalMatches', 'sum'),
        
        # Count unique players for the filter
        PlayerCount=('playerId', 'nunique')
    )
    
    # 5. Calculate the final Weighted Average
    # (Sum of all WeightedDiffs) / (Sum of all TotalMatches)
    # Use .fillna(0) in case TotalCountryMatches is 0 to avoid DivisionByZeroError
    country_stats_df['WeightedAvg_WinRate_Diff'] = (
        country_stats_df['TotalWeightedDiff'] / country_stats_df['TotalCountryMatches']
    ).fillna(0)
    
    # --- End of Calculation ---
    
    
    # 6. Apply the minimum player filter
    if min_players > 1:
        country_stats_df = country_stats_df[country_stats_df['PlayerCount'] >= min_players]
        
    # 7. Sort by the weighted average difference (descending)
    country_stats_df = country_stats_df.sort_values(by='WeightedAvg_WinRate_Diff', ascending=False)
    
    # 8. Clean up for the final report
    country_stats_df['WeightedAvg_WinRate_Diff'] = country_stats_df['WeightedAvg_WinRate_Diff'].round(2)
    
    # Return the clean, final table, dropping the intermediate calculation column
    return country_stats_df.drop(columns=['TotalWeightedDiff'])
```
